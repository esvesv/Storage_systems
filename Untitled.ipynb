```python
import sqlite3
from sqlite3 import Error
```

```python
def sql_connection():
    try:
        con = sqlite3.connect('DB_name.db')
        return con
    except Error:
        print(Error)
        
def sql_create_table(con):
    text_sql_create_table = '''
                                CREATE TABLE Students(
                                                        id integer PRIMARY KEY,
                                                        name text,
                                                        department text
                                                     )
                            '''
    cursObj = con.cursor()
    cursObj.execute(text_sql_create_table)
    con.commit()
con = sql_connection()
sql_create_table(con)
```

```python
def sql_insert(con, entities):
    text_sql_insert_into_table = '''
                                     INSERT INTO Students(
                                                             id,
                                                             name,
                                                             department
                                                          )
                                     VALUES(?, ?, ?)
                                 '''
    cursorObj = con.cursor()
    cursorObj.execute(text_sql_insert_into_table, entities)
    con.commit()
    
entities = (1, 'Jonh', 'IT')
sql_insert(con, entities)
```
